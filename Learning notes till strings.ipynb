> # **Variables**

Variables are nothing but place holders to store the Literal values. Variables can be accessed by their names and some rules have to be followed while naming variables.
 
Naming conventions for variables:
- A variable name can only contain alphabets, numbers and under score
- A variable name can only start with an underscore or an alphabet
- A variable name can't be the same as any keyword(keywords are sequences of characters which have some special meaning to them)
- A variable name can't have any spaces between the various parts of the variable. If we want to distinguish them, we can use underscore or make the variable name CamelCase

> variable_name
> VariableName

- It is suggested that we make our variables names as descriptive as possible
> a =21
> age = 21

Storing a literal into a variable is called as assigning the variable with a literal and is accomplished using the equality (=) symbol

> **Identifiers**:

- Naming a variable in a particular programming language is called an **Identifier**. 

> **Literals**

- The value we are assigning to a variable is called a **Literal**.
- A variable can hold differrent types of literals at different types.
> Ex: var=5(5 is a literal)

# Major Data Types of Variable

- The major data types of variables are:
> - int        Ex: 1,2,3,.....
> - float      Ex: 3.0,12.45.....
> - boolean    True/False
> - string     Ex: '123,'sun'...

- "None" is an empty variable.
- Any input taken by a user is always a string by default. 
> Ex: input("enter a value: ") if we enter integer value also it will be recognised as string by python.

# Type Casting

> - When we want to change values of a particular type of variable we will use this Type casting.

> Ex: Variable       Wrong value        Correct value
>---------------
>        Age         "123"(string)        int/float
>       Gender         1(int)              string
>       Count          1.5(float)           int 
> - changing string to int: int(var)
> - changing int to string: str(var)
> - changing int to float : float(var)
> Like that we can type cast from one type to another.

# Strings

## What is String

- A string is a sequence of characters.
- Computers do not deal with characters, they deal with numbers(binary). Even though you may see characters on your screen, internally it is stored and manipulated as a combination of 0's and 1's.
- This conversion of character to a number is called encoding, and the reverse process is decoding. ASCII and Unicode are some of the popular encoding used.


## Strings are Arrays

Like many other popular programming languages, strings in Python are arrays of bytesrepresenting unicode characters.
However, Python does not have a character data type, a single character is simply a string with a length of 1. Square brackets can be used to access elemnts of the string.
