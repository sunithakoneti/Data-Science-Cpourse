#### Write a Python program to display all Bivariate plots from the dataframe.

1.   Numeric to Numeric based plots
2.   Numeric to Categorical Plots
3.   Categorical to Categorical Plots






```python
### Numeric to Numeric based Plots
df_diamonds.plot.area(x="carat",y="depth",color="r")
```

```python
fig = sns.scatterplot(data = df_diamonds,x="carat",y="price",hue="cut")
plt.show()
```

```python
### Numerical vs categorical
fig =sns.violinplot(x=df_diamonds["price"],y=df_diamonds["cut"])
```

```python
sns.boxplot(x=df_diamonds["cut"],y=df_diamonds["price"])
```

```python
### categorical vs categorical
sns.stripplot(x=df_diamonds["cut"],y=df_diamonds["color"])
```
