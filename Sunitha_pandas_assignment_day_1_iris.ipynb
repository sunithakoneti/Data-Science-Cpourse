##### 2. Write a Pandas program to find the sum, mean, max, min value of 'SepalLengthCm' column of dataframe

```python
### Write a Pandas program to find the sum, mean, max, min value of 'SepalLengthCm' column of dataframe
import pandas as pd

# Read in the CSV file
df = pd.read_csv('Iris.csv')
col_sum = df["SepalLengthCm"].sum()
col_mean = df["SepalLengthCm"].mean()
col_max = df["SepalLengthCm"].max()
col_min = df["SepalLengthCm"].min()
print("The sum of 'SepalLengthCm' column is ",col_sum)
print("The mean of 'SepalLengthCm' column is ",col_mean)
print("The maximum value of 'SepalLengthCm' column is ",col_max)
print("The minimum 'SepalLengthCm' column is ",col_min)
```

##### 5. Write a Pandas program to create a subtotal of "PetalLengthCm" against Species from the dataframe.

```python
import pandas as pd
```

```python
# Method 1
df = pd.read_csv("Iris.csv")
# Extracting the data according to types of species.
df1 = df[df["Species"]=="Iris-setosa"]
df2 = df[df["Species"]=="Iris-virginica"]
df3 = df[df["Species"]== "Iris-versicolor"]
# Finding subtotal of column "PetalLengthCm" for each type of species.
sum_setosa = df1["PetalLengthCm"].sum()
sum_virginica = df2["PetalLengthCm"].sum()
sum_versicolor = df3["PetalLengthCm"].sum()
# Printing The sum of each column against species type.
print("Subtotal of 'PetalLengthCm' against Species :")
print("Sub total of Iris-setosa :",sum_setosa)
print("Sub total of Iris-virginica :",sum_virginica)
print("Sub total of Iris-versicolor :",sum_versicolor)
```

```python
# Method 2
df = pd.read_csv("Iris.csv")
# Using groupby() extracting "PetalLengthCm" column's data against Species and finding subtotalby using sum().
subtotal = df.groupby('Species')['PetalLengthCm'].sum()
print(subtotal)
```
