###### 10. Write a Pandas program to calculate the mean of price for each cut  and find maximum top 3 of diamonds DataFrame. 

```python
df = pd.read_csv('diamonds.csv')

# calculate the mean price for each cut
mean = df.groupby('cut')['price'].mean()

# sort the mean prices in descending order
mean_sort_price = mean.sort_values(ascending=False)

# find the top 3 diamonds based on their prices
top3_diamonds = df.nlargest(3, 'price')

# displaying the result
print(mean)
print(mean_sort_price)

top3_diamonds
```

###### 14. Write a Pandas program to calculate the multiply of x, y and z for each cut of diamonds DataFrame.

```python
df = pd.read_csv("diamonds.csv")
product = lambda group: group['x'] * group['y'] * group['z']

# group the DataFrame by the 'cut' column and apply the lambda function to each group
grouped = df.groupby('cut').apply(product)

grouped
```

###### 19. Write a Pandas program to read the diamonds DataFrame and detect duplicate color.

```python

# read diamonds.csv file into a DataFrame
df = pd.read_csv('diamonds.csv')

# create a boolean mask indicating whether each row is a duplicate based on the "color" column
df_duplicate = df.duplicated(subset=['color'], keep=False)

# get the rows with duplicate colors
df_duplicate_colors = df[df_duplicate]

df_duplicate_colors
```

###### 20. Write a Pandas program to count the duplicate rows of diamonds DataFrame.

```python
# read diamonds.csv file into a DataFrame
df = pd.read_csv('diamonds.csv')

# create a boolean mask indicating whether each row is a duplicate based on the "color" column
df_duplicate = df.duplicated(subset=['color'], keep=False)

duplicate_count = df_duplicate.sum()
print(duplicate_count)
```
