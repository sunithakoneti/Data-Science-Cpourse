```python
# Creating a function to display the information in Database.
tele_dict = { 'suni':'9985784934',
             'ram' : '9976853241',
             'lekha' : '9967564531',
             'gopi' : '9963788875'         
}
# Displaying The data a particulatr format
# 1. Display data
def display():
    print(" ****************************  Telephone Numbers ********************************* ")
    sorted_keys = sorted(tele_dict.keys())
    for key in sorted_keys:
        print(f"Name : {key}        Number:{tele_dict[key]} ")
   
        
# creating a function to add the data to database
# 2. Add data function
def add_contact():
    name = input("Enter Name: ")
    if name.isalpha():        
        contact = input("Enter contact number : ")
        if contact.isdigit() and len(contact) == 10 :        
            if name.lower() not in tele_dict: # if yes add the input given by user to database 
                new_entry = input("Invalid username or password. Do you want to add this user to the database (yes/no)? ")
                if new_entry.lower() == 'yes':
                    tele_dict[name] = contact
                    print("The Updated database: ")
                    display()
                else:
                    print("The access is denied")
            else:
                print("The user is already existed")
        else:
            print("Improper Contact number")
    else:
        print("Name should be an alphabet:")
# 3.Search Function definition
def search_contact():
    name1 = input("Enter the name to search: ")
    if name1.lower() in tele_dict:
        print(f"The contact number for {name1} is: {tele_dict[name1]}")
    else:
        print(f"{name1} not found in the database")
        
# Main Program
while True:
    num = input("Enter a number from 1 to 4 : ")
    print("1. Display Phone Numbers\n2. Add A Phone Number\n3. Search Contact\n4.Quit ")
    if num == '4':
        print("Quit")
        break
    elif num == '1':
        display()
    elif num == '2' :
        add_contact()
    elif num == '3':
        search_contact()
    else:
        print("Wrong number Please give a single number from 1 to 4 only! ")
```
