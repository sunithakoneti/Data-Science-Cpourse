#### 1. create a bar plot for count the frequency of the three species.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as scn
df_iris = pd.read_csv("Iris.csv")
df_iris["Species"].value_counts().plot.bar()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df_iris = pd.read_csv("Iris.csv")
a = df_iris["Species"].value_counts()
colors = sns.color_palette("husl", len(a))[0:3]
plt.xlabel("Species")
plt.ylabel("Species Values")
# Create bar plot and assign colors
plt.bar(range(len(a)), a, color=colors)

plt.show()
```

##### 2. Create a pie chart to show the percentage of species.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as scn
df_iris = pd.read_csv("Iris.csv")
df_iris["Species"].value_counts().plot.pie(autopct="%.2f%%")
```

##### 3. Create the joint to all numberical variable for analyze the relationship between two variables and describe their individual distributions on the same plot.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df_iris = pd.read_csv("Iris.csv")
df_iris["SepalLengthCm"].corr(df_iris["SepalWidthCm"])

# create joint plot
sns.jointplot(x="SepalLengthCm",y="SepalWidthCm",data=df_iris)

# show plot
plt.show()
```

##### 4. Create a scatter plot and take hue = "Species"

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df_iris = pd.read_csv("Iris.csv")
sns.scatterplot(data=df_iris, x="SepalLengthCm", y="SepalWidthCm", hue="Species")
plt.show()
```

##### 8. Create a pair chart.

##### 9. Create a Heat Map

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df_iris = pd.read_csv("Iris.csv")
# Choose a color palette
colors = sns.color_palette("husl", n_colors=5)

# Create the heatmap with the colors
sns.heatmap(df_iris.corr(), cmap=colors, annot=True)
```

###### 10. Create a distribution plot for all numerical colomns.

```python
import pandas as pd
import seaborn as sns

df_iris = pd.read_csv("Iris.csv")
df = df_iris.drop("Species", axis=1)

# Get the numerical columns
num_cols = df.select_dtypes(include=['float64', 'int64']).columns.tolist()

# Plot the distribution plot for each numerical column
for col in num_cols:
    sns.distplot(df[col])
```
