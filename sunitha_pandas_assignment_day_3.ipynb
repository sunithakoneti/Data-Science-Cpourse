###### 17. Write a Pandas program to get the current date, oldest date and number of days between Current date and oldest date of Ufo dataset.

```python
import warnings
warnings.filterwarnings('ignore')
```

```python

import pandas as pd

# Load the UFO dataset
df = pd.read_csv("ufo_sighting_data.csv")


# Convert the 'Date_time' column to datetime format
df['Date_time'] = pd.to_datetime(df['Date_time'], errors='coerce')

# Get the current date, oldest date and number of days between them
current_date = pd.Timestamp.now().normalize()
oldest_date = df['Date_time'].min().normalize()
num_days = (current_date - oldest_date).days

# Print the results
print("Current Date: ", current_date)
print("Oldest Date: ", oldest_date)
print("Number of Days: ", num_days)
```

###### 18. Write a Pandas program to get all the info of the dataframe between 1950-10-10 and 1960-10-10.

```python
import pandas as pd

# read the UFO dataset
df = pd.read_csv("ufo_sighting_data.csv")

# convert the Date_time column to datetime format
df["Date_time"] = pd.to_datetime(df["Date_time"],errors='coerce')

# filter the dataframe to include only rows between 1950-10-10 and 1960-10-10
df_filtered = df[(df["Date_time"] >= "1950-10-10") & (df["Date_time"] <= "1960-10-10")]

# print the filtered dataframe
df_filtered
```
