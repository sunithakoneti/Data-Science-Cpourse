##### 1. Write a Python program to display a bar chart of the fare of class.

```python
df_titanic = pd.read_csv("titanic.csv")
a = df_titanic["fare"].value_counts()
colors = sns.color_palette("husl",len(a))[0:3]
plt.xlabel("Fare")
plt.ylabel("Fare Values")
# Create bar plot and assign colors
plt.bar(range(len(a)),a, color=colors)

plt.show()
```

###### 2. Write a Python program to display a horizontal bar chart of the fare of class.

```python
df_titanic = pd.read_csv("titanic.csv")


# Get the fare and class data and count the values

class_counts = df_titanic["class"].value_counts()

# Plot the horizontal bar chart
plt.barh(class_counts.index, class_counts.values)
plt.title('Titanic Fare of Class Horizontal Bar Chart')
plt.xlabel('Fare')
plt.ylabel('Class')
plt.yticks(rotation=45)
plt.show()
```

###### 4. Write a Python program to create a pie chart of the fare of class.

```python
df_titanic=pd.read_csv("titanic.csv")
# Get the class data and count the values
class_counts = df_titanic["class"].value_counts()

# Define the colors for each bar
colors = sns.color_palette("husl", len(class_counts))

# Plot the bar chart with different colors
plt.pie(class_counts.values,labels=class_counts.index,  colors=colors,autopct="%.2f%%")

# Set the title and axis labels
plt.title('Fare by Class')
```

######5. Write a Python program to display a count plot of all categorical columns.

```python
df_titanic = pd.read_csv("titanic.csv")


# Get the categorical columns
cat_cols = df_titanic.select_dtypes(include=['object']).columns.tolist()

# Plot the count plot for each categorical column
for col in cat_cols:
    sns.countplot(x=col, data=df_titanic)
```

###### 6. Write a Python program to display a factor plot of ( pclass vs embarked). hist (kind = 'count')

###### 9. Write a Python program to display distribution plot for all numberical columns.

```python
df_titanic = pd.read_csv("titanic.csv")
# Get the numerical columns
num_cols = df_titanic.select_dtypes(include=['float64','int64']).columns.tolist()

# Plot the distribution plot for each numerical column
for col in num_cols:
    sns.distplot(df_titanic[col], kde=True)
```

##### Bonus Question

##### 1. What are the plots we can create on numerical columns.

### Plots for Numerical Columns:

- Histograms
- Density plots
- Box plots
- Violin plots
- Swarm plots
- Scatter plots
- Line plots

###### 2. What are the plots we can create on categorical columns.

### Plots for Categorical Columns:

- Bar plots
- Count plots
- Pie charts
- Donut charts
- Stacked bar charts
- Grouped bar charts
- Heatmaps
- Clustered bar charts

#### Note:
- Some of these plots can be used for both numerical and categorical columns, but they are generally more effective for one or the other. For example, a histogram is a good way to visualize the distribution of a numerical variable, while a bar plot is a good way to visualize the counts or frequencies of categorical variables.

- We have to Keep in mind that the choice of plot depends on the type of data and the research question being investigated. It's always a good idea to explore different types of plots to find the one that best communicates the insights you want to convey.
